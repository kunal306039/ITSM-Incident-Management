# Data Processing

* *CI_Cat*, *CI_Subcategory* had a lot of values then it replace by the major ones.unknown values replaced by model values.


```python
df.CI_Cat.replace(['database'],'storage',inplace=True)
df.CI_Cat.replace(['applicationcomponent'],'subapplication',inplace=True)
df.CI_Cat.replace(['displaydevice','officeelectronics','Phone','networkcomponents'],'hardware',inplace=True)
df.CI_Cat.replace([np.nan],'application',inplace=True)
```

```python
df.CI_Cat.isnull().sum()
```

```python
df.CI_Cat.unique()
```

```python
df.CI_Subcat.unique()
```

* Group the labels and made it one labels

```python
df.CI_Subcat.replace('Desktop','Desktop Application',inplace=True)
df.CI_Subcat.replace(['Application Server','Virtual Tape Server','ESX Server','zOS Server','Neoview Server','X86 Server',
                       'Unix Server','Oracle Server','Windows Server in extern beheer','Thin Client','NonStop Server',
                       'Number','Windows Server','Linux Server',np.nan,'SharePoint Farm','Lines'],
                       'Server Based Application',inplace=True)
df.CI_Subcat.replace('RAC Service','Banking Device',inplace=True)
df.CI_Subcat.replace(['Iptelephony','Protocol','Net Device','IPtelephony','ESX Cluster','Standard Application'],
                       'Web Based Application',inplace=True)
df.CI_Subcat.replace(['VMWare','Security Software','zOS Systeem','Firewall','Database Software','VDI','Instance',
                       'MQ Queue Manager','Automation Software','Citrix','SAP','Encryption'],'System Software',inplace=True)
df.CI_Subcat.replace(['NonStop Storage','NonStop Harddisk','Tape Library','zOS Cluster','DataCenterEquipment',
                       'MigratieDummy'],'Database',inplace=True)
df.CI_Subcat.replace(['Modem','Router'],'Network Component',inplace=True)
df.CI_Subcat.replace('KVM Switches','Switch',inplace=True)
```

```python
df.head()
```

* Replace Nan Value with mode

```python
df.Priority.replace([np.nan],df.Priority.mode(),inplace=True)
df.No_of_Related_Interactions.replace([np.nan],df.No_of_Related_Interactions.mode(),inplace=True)
df.No_of_Related_Incidents.replace([np.nan],df.No_of_Related_Incidents.mode(),inplace=True)
df.No_of_Related_Changes.replace([np.nan],df.No_of_Related_Changes.mode(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

* So much Null values in Reopen_Time, Resolved_Time, Related_Change then drop this columns.

```python
df.drop(columns=['Reopen_Time','Resolved_Time','Related_Change'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['No_of_Reassignments','Handle_Time_hrs'],axis=0,inplace=True)   #drop the rows where nan value
```

# 1. Predicting Priorites

* Feature selection

Only the features which are available when the tickets arrives can be used to predict priority. Therefore, we have used only 4 predictors.

```python
X = df.loc[:,['CI_Cat','CI_Subcat','WBS','Category']]
y = df.Priority
```

```python
y
```

```python
#Label Encoding
encode = LabelEncoder()
for i in (0,1,2,3):
  X.iloc[:,i] = encode.fit_transform(X.iloc[:,i])
```

```python
X
```

# Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Standeriztion
std = StandardScaler()
X_train = std.fit_transform(X_train)
X_test = std.transform(X_test)
```

```python
X_train
```

# Model

# 1.SVM

```python
# Train this Model
rbf_svc = SVC(kernel='rbf',C=10,gamma=0.1)
rbf_svc.fit(X_train,y_train)
```

```python
#Predict the model
y_predict_svm = rbf_svc.predict(X_test)
```

```python
print("Accuracy of SVM : ",accuracy_score(y_test,y_predict_svm))
```

```python
print("Confusion Matrix of SVM : \n",confusion_matrix(y_test,y_predict_svm))
```

```python
#Finding accuracy,Precision,Recall,F1_score
print("classification_report of SVM : \n",classification_report(y_test,y_predict_svm))
```

```python
confusion_matrix(y_test,y_predict_svm)
```

# 2. Descision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
dt_pred = dt.predict(X_test)
```

```python
#Finding Accuracy,Prescison,Recall and ConfusionMatrx
print("Accuracy Using Descison Tree : ",accuracy_score(y_test,dt_pred))
print('\n')
print(classification_report(y_test,dt_pred))
```

# Models

```python
dt_pred = dt.predict(X1_test)
```

```python
#Finding Accuracy,Prescison,Recall and ConfusionMatrx
print("Accuracy Using Descison Tree : ",accuracy_score(y1_test,dt_pred))
print('\n')
print(classification_report(y1_test,dt_pred))
```
